```python
import gym
from gym_chess import ChessEnvV1
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
# Create Chess environment
def make_env():
    return ChessEnvV1(player_color="WHITE", opponent="random", log=True)

# Vectorized environment
env = DummyVecEnv([make_env])
```

```python
# Initialize PPO model
model = PPO("MlpPolicy", env, verbose=0, tensorboard_log="./ppo_chess_tensorboard/")

# Train the model and collect loss values
TIMESTEPS = 50000
losses = []

def callback(_locals, _globals):
    if 'loss' in _locals:
        losses.append(_locals['loss'])
    return True

model.learn(total_timesteps=TIMESTEPS, callback=callback, progress_bar=True)

# Save the model
model.save("ppo_chess_v1")
```

```python
import matplotlib.pyplot as plt

# Plot the loss evolution
plt.plot(losses)
plt.xlabel("Training Step")
plt.ylabel("Loss")
plt.title("PPO Training Loss Evolution")
plt.show()
```

```python
# Load the trained model
model = PPO.load("ppo_chess_v1")

# Evaluate the model
def evaluate_agent(model, env, num_episodes=10):
    rewards = []
    for _ in range(num_episodes):
        obs = env.reset()
        done = False
        total_reward = 0
        while not done:
            action, _states = model.predict(obs, deterministic=True)
            obs, reward, done, info = env.step(action)
            total_reward += reward
        rewards.append(total_reward)
    return rewards

# Evaluate against a random opponent
eval_rewards = evaluate_agent(model, env, num_episodes=10)
print(f"Evaluation rewards over 10 games: {eval_rewards}")
```

```python
# Create Chess environment
def make_env():
    return ChessEnvV1(player_color="WHITE", opponent="random", log=True)

# Vectorized environment
env = DummyVecEnv([make_env])

# Play a single game and print the moves
def play_game(model, env):
    obs = env.reset()
    done = False
    while not done:
        print(env.render())
        action, _states = model.predict(obs, deterministic=True)
        obs, reward, done, info = env.step(action)
        print(reward)
    print(env.render())

print("Playing a game with the trained model:")
play_game(model, env)
```
